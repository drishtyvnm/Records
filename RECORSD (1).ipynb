This repository depicts my ***semester wise academic grade records***. The data is visualized and organized programmatically to reflect my academic progress.

```python
import matplotlib.pyplot as plt
x=["Sem-1","Sem-2", "Sem-3", "Sem-4", "Sem-5","Sem-6","Sem-7","Sem-8"]
y=[86,84,80,82,87,90,88,96]
colors=[10,78,49,89,56,70,29,30]
sizes=[200,250,600,500,550,400,450,540]

plt.title("Records", fontsize=15)
plt.xlabel("Semesters", fontsize=15)
plt.ylabel("Marks", fontsize=15)
plt.scatter(x,y,c=colors,s=sizes,cmap="viridis", alpha=0.5,marker= '*', edgecolor='r',linestyle=':')
plt.show()
```
